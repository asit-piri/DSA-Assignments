# Question 1

ðŸ’¡ Convert 1D Array Into 2D Array

You are given a **0-indexed** 1-dimensional (1D) integer array original, and two integers, m and n. You are tasked with creating a 2-dimensional (2D) array with Â m rows and n columns using **all** the elements from original.

The elements from indices 0 to n - 1 (**inclusive**) of original should form the first row of the constructed 2D array, the elements from indices n to 2 * n - 1 (**inclusive**) should form the second row of the constructed 2D array, and so on.

Return *an* m x n *2D array constructed according to the above procedure, or an empty 2D array if it is impossible*.

**Example 1:**

image.png

**Input:** original = [1,2,3,4], m = 2, n = 2

**Output:** [[1,2],[3,4]]

**Explanation:** The constructed 2D array should contain 2 rows and 2 columns.

The first group of n=2 elements in original, [1,2], becomes the first row in the constructed 2D array.

The second group of n=2 elements in original, [3,4], becomes the second row in the constructed 2D array.

## Solution

To solve this problem, we can iterate over the elements of the original array and construct the 2D array based on the given conditions. We'll check if it's possible to construct the 2D array by comparing the size of the original array with the product of m and n.

Here's the step-by-step approach to solve the problem:

1. Check if the size of the original array is equal to m * n. If not, return an empty 2D array since it's impossible to construct the desired array.

2. Create an empty 2D array result with m rows and n columns.

3. Initialize two variables, row and col, to 0. These will keep track of the current row and column in the result array.

4. Iterate over the elements of the original array.

    - Assign the current element to result[row][col].
    
    - Increment col by 1.

    - If col becomes equal to n, reset col to 0 and increment row by 1.
5. Return the result array.

Here's the Python code that implements this approach:

```python
def construct2DArray(original, m, n):
    if len(original) != m * n:
        return []

    result = [[0] * n for _ in range(m)]
    row, col = 0, 0

    for num in original:
        result[row][col] = num
        col += 1

        if col == n:
            col = 0
            row += 1

    return result
```

```python
# Test Case 2

original = [7, 4, 8, 9]
m = 2
n = 2
result = construct2DArray(original, m, n)
print(result)
```

### Conclusion

The construct2DArray function takes the original array, m, and n as input. It first checks if the size of the original array matches the required size for the 2D array (m * n). If not, it returns an empty array since it's impossible to construct the desired array.

If the size is valid, it creates an empty 2D array result with m rows and n columns. It then iterates over the elements of the original array and assigns them to the corresponding positions in the result array. After assigning an element, it updates the row and col indices accordingly.

Finally, it returns the constructed result array.

The time complexity of this approach is O(m * n) since we iterate over the original array once and construct the 2D array. The space complexity is also O(m * n) since we create the result array of size m rows and n columns.

ðŸ’¡ You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**

image.png

**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.

## Solution

To solve this problem, we can use a binary search approach. Since the number of coins required to build each row follows a pattern (1, 2, 3, ...), we can find the maximum number of complete rows by performing a binary search on the number of rows.

Here's the step-by-step approach to solve the problem:

1. Initialize two variables, left and right, to 1 and n respectively. These variables represent the range of possible numbers of rows.

2. Perform a binary search while left is less than or equal to right.

    - Calculate the mid point as mid = left + (right - left) // 2.

    - Calculate the total number of coins required to build mid rows as total = (mid * (mid + 1)) // 2.

     - If total is less than or equal to n, update left = mid + 1 since we can build mid complete rows. Otherwise, update right = mid - 1.

3. After the binary search, the value of right will be the number of complete rows that can be built.

Here's the Python code that implements this approach:

```python
def arrangeCoins(n):
    left, right = 1, n

    while left <= right:
        mid = left + (right - left) // 2
        total = (mid * (mid + 1)) // 2

        if total <= n:
            left = mid + 1
        else:
            right = mid - 1

    return right
```

### Conclusion

The arrangeCoins function takes the integer n as input and performs a binary search to find the number of complete rows that can be built. It initializes left and right to 1 and n respectively, representing the range of possible numbers of rows.

In each iteration of the binary search, it calculates the mid point and the total number of coins required to build mid rows. If the total is less than or equal to n, it updates left to mid + 1 since we can build mid complete rows. Otherwise, it updates right to mid - 1.

After the binary search, the value of right represents the number of complete rows that can be built, so it is returned as the result.

The **time complexity of this approach is O(log n)** since we perform a binary search on the range of possible numbers of rows. The **space complexity is O(1)** as we only use a constant amount of extra space.

# Question 3

ðŸ’¡ Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

**Input:** nums = [-4,-1,0,3,10]

**Output:** [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].

After sorting, it becomes [0,1,9,16,100].

## Solution

To solve this problem, we can follow a two-pointer approach. Since the given array nums is already sorted in non-decreasing order, we can compare the squares of the numbers at the two ends of the array to determine the order of the squared values.

Here's the step-by-step approach to solve the problem:

1. Initialize two pointers, left and right, to point to the start and end of the nums array.

2. Create an empty result array to store the squared values.

3. While the left pointer is less than or equal to the right pointer:

    - Compare the absolute values of nums[left] and nums[right].

    - If the square of nums[left] is greater than or equal to the square of nums[right]:
    
    - Append the square of nums[left] to the result array.

    - Move the left pointer to the next position.

    - Otherwise, if the square of nums[left] is less than the square of nums[right]:

    - Append the square of nums[right] to the result array.

    - Move the right pointer to the previous position.

4. Reverse the result array to obtain the squares in non-decreasing order.

5. Return the result array.

Here's the Python code that implements this approach:

```python
def sortedSquares(nums):
    left, right = 0, len(nums) - 1
    result = []

    while left <= right:
        if abs(nums[left]) >= abs(nums[right]):
            result.append(nums[left] ** 2)
            left += 1
        else:
            result.append(nums[right] ** 2)
            right -= 1

    return result[::-1]
```

### Test Cases

```python
# Test the function
nums = [-4, -1, 0, 3, 10]
result = sortedSquares(nums)
print(result)
```

### Conclusion

The sortedSquares function takes the array nums as input and follows the two-pointer approach to compute the squares of the numbers and store them in the result array.

In each iteration, the function compares the absolute values of the numbers at the left and right pointers. If the square of the number at the left pointer is greater than or equal to the square of the number at the right pointer, it appends the square of the number at the left pointer to the result array and moves the left pointer to the next position. Otherwise, it appends the square of the number at the right pointer to the result array and moves the right pointer to the previous position.

After the loop, the result array contains the squared values in reverse order, so we reverse the array using slicing ([::-1]) to obtain the squares in non-decreasing order.

The **time complexity of this approach is O(n)**, where n is the length of the input array nums, as we iterate through the array once. The **space complexity is O(n)** as we create a result array of the same length as nums to store the squared values.

# Question 4

ðŸ’¡ Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

## Solution

To solve this problem, we can use sets to efficiently determine the distinct integers that are present in one array but not in the other. We'll iterate through both arrays and add the elements to their respective sets. Then, we'll find the set differences to obtain the distinct integers that are not present in the other array.

Here's the step-by-step approach to solve the problem:

1. Create two empty sets, set1 and set2, to store the distinct integers from nums1 and nums2, respectively.

2. Iterate through nums1 and add each element to set1.

3. Iterate through nums2 and add each element to set2.

4. Find the set difference diff1 between set1 and set2 to obtain the distinct 

5. integers in nums1 that are not present in nums2.

6. Find the set difference diff2 between set2 and set1 to obtain the distinct 

7. integers in nums2 that are not present in nums1.

8. Create a list answer and set answer[0] to diff1 and answer[1] to diff2.

9. Return answer.

Here's the Python code that implements this approach:

```python
def findDisappearedNumbers(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    diff1 = list(set1 - set2)
    diff2 = list(set2 - set1)
    answer = [diff1, diff2]
    return answer
```

### Test Cases

```python
# Test Case 1

nums1 = [1, 2, 3]
nums2 = [2, 4, 6]
answer = findDisappearedNumbers(nums1, nums2)
print(answer)
```

### Conclusion

The findDisappearedNumbers function takes nums1 and nums2 as input arrays. It creates two sets, set1 and set2, and adds the elements from nums1 and nums2 to their respective sets. Then, it finds the set differences between set1 and set2 to obtain the distinct integers in nums1 that are not present in nums2, and vice versa.

Finally, it creates the answer list with diff1 and diff2 and returns it.

The **time complexity of this approach is O(m + n)**, where m and n are the lengths of nums1 and nums2, respectively. The **space complexity is O(m + n**) as we create sets and lists to store the distinct integers.

# Question 5

ðŸ’¡ Given two integer arrays arr1 and arr2, and the integer d, *return the distance value between the two arrays*.

The distance value is defined as the number of elements arr1[i] such that there is not any element arr2[j] where |arr1[i]-arr2[j]| <= d.

**Example 1:**

**Input:** arr1 = [4,5,8], arr2 = [10,9,1,8], d = 2

**Output:** 2

**Explanation:**

For arr1[0]=4 we have:

|4-10|=6 > d=2

|4-9|=5 > d=2

|4-1|=3 > d=2

|4-8|=4 > d=2

For arr1[1]=5 we have:

|5-10|=5 > d=2

|5-9|=4 > d=2

|5-1|=4 > d=2

|5-8|=3 > d=2

For arr1[2]=8 we have:

**|8-10|=2 <= d=2**

**|8-9|=1 <= d=2**

|8-1|=7 > d=2

**|8-8|=0 <= d=2**


## Solution

To solve this problem, we can iterate through each element in arr1 and check if there is any element in arr2 that satisfies the condition |arr1[i] - arr2[j]| <= d. If no such element is found, we increment a counter variable. Finally, we return the value of the counter variable as the distance value.

Here's the step-by-step approach to solve the problem:

1. Initialize a counter variable distance to 0.

2. Iterate through each element num1 in arr1.

    - For each num1, iterate through each element num2 in arr2.

    - If |num1 - num2| is less than or equal to d, break out of the inner loop.

    - If the inner loop completes without finding a satisfying element, increment distance by 1.

3. Return the value of distance as the distance value.

Here's the Python code that implements this approach:

```python
def findTheDistanceValue(arr1, arr2, d):
    distance = 0
    for num1 in arr1:
        for num2 in arr2:
            if abs(num1 - num2) <= d:
                break
        else:
            distance += 1
    return distance
```

### Conclusion

The findTheDistanceValue function takes arr1, arr2, and d as input. It initializes the distance variable to 0 and then iterates through each element in arr1. For each element in arr1, it iterates through each element in arr2 and checks if the absolute difference between them is less than or equal to d. If a satisfying element is found, it breaks out of the inner loop. If the inner loop completes without finding a satisfying element, it increments the distance variable by 1.

Finally, it returns the value of distance as the distance value.

The time complexity of this approach is O(n * m), where n is the length of arr1 and m is the length of arr2. The space complexity is O(1) as we only use a constant amount of extra space for variables.

# Question 6

Given an integer array nums of length n where all the integers of nums are in the range [1, n] and each integer appears **once** or **twice**, return *an array of all the integers that appears **twice***.

You must write an algorithm that runs in O(n) time and uses only constant extra space.

**Example 1:**

**Input:** nums = [4,3,2,7,8,2,3,1]

**Output:**

[2,3]

## Solution

To find the integers that appear twice in the given array nums, we can utilize the fact that the integers are in the range [1, n] and each integer appears once or twice.

Here's an algorithm that runs in O(n) time and uses only constant extra space:

1. Initialize an empty list result to store the integers that appear twice.

2. Iterate through each element num in nums.

    - Calculate the index idx as abs(num) - 1.

    - If nums[idx] is negative, it means we have encountered num before. Append abs(num) to result.

    - Otherwise, negate nums[idx] to mark the element as visited.

3. Return result.

Here's the Python code that implements this algorithm:

```python
def findDuplicates(nums):
    result = []
    for num in nums:
        idx = abs(num) - 1
        if nums[idx] < 0:
            result.append(abs(num))
        else:
            nums[idx] = -nums[idx]
    return result
```

### Test Cases

```python
# Test Case 1

nums = [4, 3, 2, 7, 8, 2, 3, 1]
result = findDuplicates(nums)
print(result)
# Output: [2, 3]
```

### Conclusion

The findDuplicates function takes nums as input and initializes an empty list result. It then iterates through each element in nums. For each element, it calculates the corresponding index idx and checks if nums[idx] is negative. If it is negative, it means we have encountered the element before and it is a duplicate, so we append the absolute value of the element to result. Otherwise, it negates nums[idx] to mark the element as visited.

Finally, it returns the result list containing the integers that appear twice.

The **time complexity of this algorithm is O(n)** since we iterate through each element in nums exactly once. The **space complexity is O(1)** as we only use a constant amount of extra space for variables.

# Question 7

ðŸ’¡ Suppose an array of length n sorted in ascending order is **rotated** between 1 and n times. For example, the array nums = [0,1,2,4,5,6,7] might become:

- [4,5,6,7,0,1,2] if it was rotated 4 times.
- [0,1,2,4,5,6,7] if it was rotated 7 times.

Notice that **rotating** an array [a[0], a[1], a[2], ..., a[n-1]] 1 time results in the array [a[n-1], a[0], a[1], a[2], ..., a[n-2]].

Given the sorted rotated array nums of **unique** elements, return *the minimum element of this array*.

You must write an algorithm that runs in O(log n) time.

**Example 1:**

**Input:** nums = [3,4,5,1,2]

**Output:** 1

**Explanation:**

The original array was [1,2,3,4,5] rotated 3 times.

## Solution

To find the minimum element in a sorted rotated array with unique elements, we can use a modified version of binary search.

Here's an algorithm that runs in O(log n) time:

1. Initialize two pointers, left and right, pointing to the start and end of the array respectively.

2. While left is less than right:

    - Calculate the middle index as mid using the formula: mid = left + (right - left) // 2.

    - Check if nums[mid] is greater than nums[right]. If true, it means the minimum element lies in the right half of the array. Set left = mid + 1.

    - Otherwise, the minimum element lies in the left half of the array or is equal to nums[mid]. Set right = mid.

3. At the end of the loop, left will be pointing to the minimum element.

4. Return nums[left] as the result.

Here's the Python code that implements this algorithm:

```python
def findMin(nums):
    left = 0
    right = len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] > nums[right]:
            left = mid + 1
        else:
            right = mid

    return nums[left]
```

```python
# Test Case 2

nums = [3, 4, 5, 1, 2, 0]
min_element = findMin(nums)
print(min_element)
```

### Conclusion

The findMin function takes nums as input and initializes the left pointer to 0 and the right pointer to the index of the last element in the array. It then enters a while loop that continues until left is less than right.

Inside the loop, it calculates the middle index mid and compares nums[mid] with nums[right]. If nums[mid] is greater than nums[right], it means the minimum element lies in the right half of the array, so it updates left to mid + 1. Otherwise, the minimum element lies in the left half of the array or is equal to nums[mid], so it updates right to mid.

Finally, it returns nums[left] as the minimum element of the array.

The **time complexity of this algorithm is O(log n)** as it performs binary search, which reduces the search space in half at each step.

1. Initialize an empty hash set called originalSet.

2. Iterate through each element num in the changed array.

    - Check if num/2 is present in the originalSet. If it is, remove num/2 from the set.

    - Otherwise, add num to the originalSet.

Here's the Python code that implements this algorithm:

```python
'''
def findOriginalArray(changed):
    if len(changed) % 2 != 0:
        return []
    
    originalSet = set()
    for num in changed:
        if num / 2 in originalSet:
            originalSet.remove(num / 2)
        else:
            originalSet.add(num)
    
    if len(originalSet) == 0:
        return []
    
    return list(originalSet)
'''

from collections import defaultdict

def findOriginalArray(changed):
    if len(changed) % 2 != 0:
        return []  # If the length is odd, it can't be a doubled array

    count = defaultdict(int)
    for num in changed:
        count[num] += 1

    original = []
    for num in changed:
        if count[num] == 0:
            continue
        if count[num // 2] > 0:
            count[num // 2] -= 1
            original.append(num // 2)
        else:
            return []

    return original
```
