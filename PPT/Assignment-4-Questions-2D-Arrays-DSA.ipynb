# Question 1

ðŸ’¡  Given three integer arrays arr1, arr2 and arr3 **sorted** in **strictly increasing** order, return a sorted array of **only** the integers that appeared in **all** three arrays.

**Example 1:**

Input: arr1 = [1,2,3,4,5], arr2 = [1,2,5,7,9], arr3 = [1,3,4,5,8]

Output: [1,5]

**Explanation:** Only 1 and 5 appeared in the three arrays.

## Solution

To find the integers that appeared in all three arrays, you can use a simple approach by iterating through the arrays simultaneously. Since the arrays are sorted, you can compare the elements at corresponding indices to determine if they are equal. 

Here's an example implementation in Python:

```python
def findCommonElements(arr1, arr2, arr3):
    result = []
    i = j = k = 0

    while i < len(arr1) and j < len(arr2) and k < len(arr3):
        if arr1[i] == arr2[j] == arr3[k]:
            result.append(arr1[i])
            i += 1
            j += 1
            k += 1
        elif arr1[i] < arr2[j]:
            i += 1
        elif arr2[j] < arr3[k]:
            j += 1
        else:
            k += 1

    return result
```

```python
# Test Case 1

arr1 = [1, 2, 3, 4, 5]
arr2 = [1, 2, 5, 7, 9]
arr3 = [1, 3, 4, 5, 8]

common_elements = findCommonElements(arr1, arr2, arr3)
print(common_elements)
```

```python
# Test Case 2

arr1 = [1, 2, 3, 4, 5, 6]
arr2 = [1, 2, 5, 6, 7, 9]
arr3 = [1, 3, 4, 5, 6, 8]

common_elements = findCommonElements(arr1, arr2, arr3)
print(common_elements)
```

The **time complexity of this solution is O(N)**, where N is the total number of elements across all three arrays. The algorithm iterates through the arrays once, comparing the elements and appending the common elements to the result.

The **space complexity is O(1)**, as the resulting array is created and returned, but it doesn't require any additional space that grows with the input size.

# Question 2

ðŸ’¡ Given two **0-indexed** integer arrays nums1 and nums2, return *a list* answer *of size* 2 *where:*

- answer[0] *is a list of all **distinct** integers in* nums1 *which are **not** present in* nums2*.*
- answer[1] *is a list of all **distinct** integers in* nums2 *which are **not** present in* nums1.

**Note** that the integers in the lists may be returned in **any** order.

**Example 1:**

**Input:** nums1 = [1,2,3], nums2 = [2,4,6]

**Output:** [[1,3],[4,6]]

**Explanation:**

For nums1, nums1[1] = 2 is present at index 0 of nums2, whereas nums1[0] = 1 and nums1[2] = 3 are not present in nums2. Therefore, answer[0] = [1,3].

For nums2, nums2[0] = 2 is present at index 1 of nums1, whereas nums2[1] = 4 and nums2[2] = 6 are not present in nums2. Therefore, answer[1] = [4,6].

## Solutions

To find the distinct integers that are present in one array but not in the other, we can use a set to efficiently perform the comparison. Here's an approach to solve this problem:

1. Initialize two sets, set1 and set2, to store the unique elements of nums1 and nums2, respectively.

2. Iterate through each element in nums1 and add it to set1.

3. Iterate through each element in nums2 and add it to set2.

4. Find the set difference between set1 and set2 to obtain the distinct integers in nums1 that are not present in nums2. Assign the result to result1.

5. Find the set difference between set2 and set1 to obtain the distinct integers in nums2 that are not present in nums1. Assign the result to result2.

6. Convert result1 and result2 to lists and return them as answer.

Here's the Python code that implements this approach:

```python
def findDisjoint(nums1, nums2):
    set1 = set(nums1)
    set2 = set(nums2)
    result1 = list(set1 - set2)
    result2 = list(set2 - set1)
    return [result1, result2]
```

The function uses sets to efficiently find the set differences and converts them back to lists before returning.

The **time complexity of this approach is O(m + n)**, where m and n are the lengths of nums1 and nums2, respectively. The **space complexity is O(m + n)** as we use sets to store the unique elements.

ðŸ’¡Given a 2D integer array matrix, return *the **transpose** of* matrix.

The **transpose** of a matrix is the matrix flipped over its main diagonal, switching the matrix's row and column indices.

**Example 1:**

Input: matrix = [[1,2,3],[4,5,6],[7,8,9]]

Output: [[1,4,7],[2,5,8],[3,6,9]]

image.png

## Solution

To find the transpose of a matrix, we need to swap the elements across the main diagonal. Here's an approach to solve this problem:

1. Initialize an empty result matrix with dimensions cols x rows, where cols is the number of columns in the original matrix and rows is the number of rows.

2. Iterate through each element in the original matrix. For each element matrix[i][j], assign it to the corresponding position in the result matrix as result[j][i].

3. Return the result matrix.

Here's the Python code that implements this approach:

```python
def transpose(matrix):
    rows = len(matrix)
    cols = len(matrix[0])
    result = [[0] * rows for _ in range(cols)]
    for i in range(rows):
        for j in range(cols):
            result[j][i] = matrix[i][j]
    return result
```

### Conclusion

The transpose function takes a matrix as input and returns the transpose of the matrix. It initializes an empty result matrix with dimensions cols x rows and then iterates through each element of the original matrix, assigning it to the corresponding position in the result matrix. Finally, it returns the result matrix.

The time complexity of this approach is O(m * n), where m is the number of rows and n is the number of columns in the original matrix. The space complexity is O(m * n) as we need to allocate a new matrix for the result.

**Example 1:**

Input: nums = [1,4,3,2]

Output: 4

**Explanation:** All possible pairings (ignoring the ordering of elements) are:

1. (1, 4), (2, 3) -> min(1, 4) + min(2, 3) = 1 + 2 = 3

2. (1, 3), (2, 4) -> min(1, 3) + min(2, 4) = 1 + 2 = 3

3. (1, 2), (3, 4) -> min(1, 2) + min(3, 4) = 1 + 3 = 4

So the maximum possible sum is 4.

## Solution

To maximize the sum of the minimum elements in each pair, we can sort the array nums in ascending order and then pair up the adjacent elements. Since the array is sorted, the minimum element in each pair will be the first element of the pair.

Here's the approach to solve this problem:

1. Sort the array nums in ascending order.

2. Initialize a variable max_sum to 0.

3. Iterate over the sorted array nums by incrementing the index by 2 in each iteration.

4. Add the minimum value between nums[i] and nums[i+1] to max_sum.

5. Return the value of max_sum.

Here's the Python code that implements this approach:

```python
def arrayPairSum(nums):
    nums.sort()
    max_sum = 0
    for i in range(0, len(nums), 2):
        max_sum += nums[i]
    return max_sum
```

ðŸ’¡ You have n coins and you want to build a staircase with these coins. The staircase consists of k rows where the ith row has exactly i coins. The last row of the staircase **may be** incomplete.

Given the integer n, return *the number of **complete rows** of the staircase you will build*.

**Example 1:**

**Input:** n = 5

**Output:** 2

**Explanation:** Because the 3rd row is incomplete, we return 2.

image.png


## Solution

To determine the number of complete rows of a staircase given a certain number of coins, we can use a binary search approach. The idea is to find the maximum number of complete rows that can be formed such that the total number of coins used does not exceed the given number of coins.

Here's the approach to solve this problem:

Initialize two pointers, left and right, to 1 and n respectively. These pointers represent the range of possible numbers of complete rows.
While left is less than or equal to right, do the following:
Calculate the midpoint as mid = left + (right - left) // 2.
Calculate the total number of coins required to form mid complete rows using the formula total = (mid * (mid + 1)) // 2.
If total is less than or equal to n, update left to mid + 1 since we can try to form more complete rows.
Otherwise, update right to mid - 1 since we need to try with a smaller number of complete rows.
Return right as the maximum number of complete rows.
Here's the Python code that implements this approach:

```python
def arrangeCoins(n):
    left, right = 1, n
    while left <= right:
        mid = left + (right - left) // 2
        total = (mid * (mid + 1)) // 2
        if total <= n:
            left = mid + 1
        else:
            right = mid - 1
    return right
```

### Test Cases

```python
# Test Case 1
n = 5
complete_rows = arrangeCoins(n)
print(complete_rows)
```

### Conclusion

The arrangeCoins function takes an integer n as input and returns the maximum number of complete rows of the staircase. It initializes the left and right pointers to 1 and n respectively. Then, it performs a binary search to find the maximum number of complete rows. The function calculates the midpoint, mid, and the total number of coins required to form mid complete rows. Based on the comparison of total with n, it updates either left or right accordingly. Finally, it returns right as the maximum number of complete rows.

The **time complexity of this approach is O(log n)** since it uses binary search to find the answer. The **space complexity is O(1)** as we are using only a constant amount of extra space.

ðŸ’¡ Given an integer array nums sorted in **non-decreasing** order, return *an array of **the squares of each number** sorted in non-decreasing order*.

**Example 1:**

Input: nums = [-4,-1,0,3,10]

Output: [0,1,9,16,100]

**Explanation:** After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100]

## Solution

To square each number in the input array and return the result in non-decreasing order, we can utilize a two-pointer approach. Since the input array is already sorted in non-decreasing order, we can compare the squares of the numbers from both ends and insert them into a new array in descending order of magnitude.

Here's the step-by-step approach to solve this problem:

Initialize two pointers, left and right, to the start and end of the input array nums.
Initialize an empty result array, squares, to store the squares of the numbers.
Iterate over the input array from the end towards the start:
Compare the squares of nums[left] and nums[right].
If the square of nums[left] is greater than or equal to the square of nums[right], append it to the front of squares and increment left by 1.
Otherwise, append the square of nums[right] to the front of squares and decrement right by 1.
Return the reversed squares array.
Here's the Python code that implements this approach:

```python
def sortedSquares(nums):
    left, right = 0, len(nums) - 1
    squares = []
    while left <= right:
        if nums[left] ** 2 >= nums[right] ** 2:
            squares.insert(0, nums[left] ** 2)
            left += 1
        else:
            squares.insert(0, nums[right] ** 2)
            right -= 1
    return squares[::-1]
```

```python
# Test Case 2
nums = [0, 3, 5, 7, 10]
result = sortedSquares(nums)
print(result)
```

### Conclusion

The sortedSquares function takes an array nums as input and returns the array of squares of each number in non-decreasing order. It initializes the left and right pointers to the start and end of the array, and the squares array as an empty list. Then, it iterates over the array from the end towards the start, compares the squares of the numbers using the two pointers, and appends the squares to the front of the squares array accordingly. Finally, it returns the reversed squares array to obtain the result in non-decreasing order.

The **time complexity of this approach is O(n)** since we iterate over the input array once. The **space complexity is also O(n)** as we use an additional array, squares, to store the squared values.

ðŸ’¡ You are given an m x n matrix M initialized with all 0's and an array of operations ops, where ops[i] = [ai, bi] means M[x][y] should be incremented by one for all 0 <= x < ai and 0 <= y < bi.

Count and return *the number of maximum integers in the matrix after performing all the operations*

**Example 1:**

**Input:** m = 3, n = 3, ops = [[2,2],[3,3]]

**Output:** 4

**Explanation:** The maximum integer in M is 2, and there are four of it in M. So return 4.

image.png

## Solution

To solve this problem, we can simulate the operations on the matrix and track the minimum dimensions of the rectangle that covers all the operations. The number of maximum integers in the resulting matrix will be equal to the area of this rectangle.

Here's the step-by-step approach to solve the problem:

Initialize min_row and min_col to m and n respectively, which represent the minimum dimensions of the rectangle covering all the operations.
Iterate over the operations in reverse order:
Update min_row as the minimum value between min_row and ops[i][0].
Update min_col as the minimum value between min_col and ops[i][1].
Return the product of min_row and min_col.
Here's the Python code that implements this approach:

```python
def maxCount(m, n, ops):
    min_row = m
    min_col = n
    for op in ops:
        min_row = min(min_row, op[0])
        min_col = min(min_col, op[1])
    return min_row * min_col
```

### Test Cases

```python
# Test Case 1
m = 3
n = 3
ops = [[2, 2], [3, 3]]
result = maxCount(m, n, ops)
print(result)
```

### Conclusion

The maxCount function takes the dimensions m and n of the matrix and the operations array ops as input. It initializes min_row and min_col to m and n respectively. Then, it iterates over the operations in reverse order and updates min_row and min_col to the minimum dimensions of the covering rectangle. Finally, it returns the product of min_row and min_col as the number of maximum integers in the resulting matrix.

The **time complexity of this approach is O(k)**, where k is the length of the ops array. We iterate over the operations once. The **space complexity is O(1)** as we use only a constant amount of additional space to track the minimum dimensions.

*Return the array in the form* [x1,y1,x2,y2,...,xn,yn].

**Example 1:**

**Input:** nums = [2,5,1,3,4,7], n = 3

**Output:** [2,3,5,4,1,7]

**Explanation:** Since x1=2, x2=5, x3=1, y1=3, y2=4, y3=7 then the answer is [2,3,5,4,1,7].

## Solution

To transform the given array nums into the desired form [x1, y1, x2, y2, ..., xn, yn], we can create a new array and iterate over nums in a specific pattern.

Here's the step-by-step approach to solve the problem:

Create an empty array result to store the transformed elements.
Iterate i from 0 to n-1:
Append nums[i] to result.
Append nums[i+n] to result.
Return the result array.
Here's the Python code that implements this approach:

```python
def shuffle(nums, n):
    result = []
    for i in range(n):
        result.append(nums[i])
        result.append(nums[i + n])
    return result
```

# Question 1

Given an m*n matrix, return all elements of the matrix in spiral order

Example:

1 -> 2 -> 3

4 -> 5    6

7 <- 8 <- 9

Input: matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]

## Solution

To return all elements of the matrix in spiral order, we can use a simulation approach. We define four variables to keep track of the boundaries of the current spiral: top, bottom, left, and right. We start with top = 0, bottom = m-1, left = 0, and right = n-1, where m and n are the number of rows and columns in the matrix.

Here's the step-by-step approach to solve the problem:

Create an empty array result to store the elements in spiral order.
Initialize the variables top, bottom, left, and right to their respective boundaries: top = 0, bottom = m-1, left = 0, right = n-1.
Run a while loop until top is less than or equal to bottom and left is less than or equal to right:
Iterate from left to right and append matrix[top][i] to result.
Increment top by 1.
Iterate from top to bottom and append matrix[i][right] to result.
Decrement right by 1.
Check if top is still less than or equal to bottom:
Iterate from right to left and append matrix[bottom][i] to result.
Decrement bottom by 1.
Check if left is still less than or equal to right:
Iterate from bottom to top and append matrix[i][left] to result.
Increment left by 1.
Return the result array.
Here's the Python code that implements this approach:

```python
def spiralOrder(matrix):
    if not matrix:
        return []
    
    result = []
    m, n = len(matrix), len(matrix[0])
    top, bottom = 0, m - 1
    left, right = 0, n - 1

    while top <= bottom and left <= right:
        for i in range(left, right + 1):
            result.append(matrix[top][i])
        top += 1

        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        if top <= bottom:
            for i in range(right, left - 1, -1):
                result.append(matrix[bottom][i])
            bottom -= 1

        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    return result
```

### Test Cases

```python
# Test Case 1
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = spiralOrder(matrix)
print(result)
# Output: [1, 2, 3, 6, 9, 8, 7, 4, 5]
```

### Conclusion

The spiralOrder function takes the matrix as input and initializes the variables top, bottom, left, and right to their respective boundaries. It then runs a while loop until the conditions are met. In each iteration, it iterates over the elements of the current spiral, appending them to the result array. Finally, it returns the result array.

The time complexity of this approach is O(m * n), where m is the number of rows and n is the number of columns in the matrix. We need to visit every element once. The space complexity is O(1) since we are not using any additional space that grows with the input size.

Given a square matrix mat, return the sum of the matrix diagonals. Only include the sum of all the elements on the primary diaginal and all the elements on the secondary diagonal that are not part of the primary diagonal.

example

1 2 3

7 8 9

Input: mat[[1,2,3], [4,5,6], [7,8,9]]

Output: 25

Explanation: Diagnols sum: 1 + 5 + 9 + 3 + 7 = 25

Notice that element mat [1][1] = 5 is cpunted only once.

## Solution

To calculate the sum of the diagonals in a square matrix, we can iterate over the elements and accumulate the values of the primary diagonal and the secondary diagonal that are not part of the primary diagonal.

Here's the step-by-step approach to solve the problem:

1. Initialize a variable n to the number of rows or columns in the matrix (assuming it is a square matrix).

2. Initialize a variable sum to 0 to store the sum of the diagonals.

3. Iterate over the elements in the matrix using a nested loop:

    - For each element, check if its indices i and j are equal (part of the primary diagonal). If they are, add the element's value to the sum.
    
    - If the indices are not equal and the sum of the indices is equal to n - 1 (part of the secondary diagonal that is not part of the primary diagonal), add the element's value to the sum.

4. Return the value of sum.

Here's the Python code that implements this approach:

```python
def diagonalSum(mat):
    n = len(mat)
    diag_sum = 0

    for i in range(n):
        for j in range(n):
            if i == j:
                diag_sum += mat[i][j]
            elif i + j == n - 1:
                diag_sum += mat[i][j]

    return diag_sum
```

### Test Cases

```python
# Test Case 1
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = diagonalSum(matrix)
print(result)
# Output: 25
```

### Conclusion

The diagonalSum function takes the matrix as input and initializes the variable n to the number of rows or columns in the matrix. It then iterates over the elements using two nested loops. For each element, it checks if the indices i and j are equal (part of the primary diagonal) and adds the value to the diag_sum. If the indices are not equal and the sum of the indices is equal to n - 1 (part of the secondary diagonal that is not part of the primary diagonal), it adds the value to the diag_sum. Finally, it returns the value of diag_sum.

The **time complexity of this approach is O(n^2)**, where n is the number of rows or columns in the matrix. We need to visit every element once. The **space complexity is O(1)** since we are not using any additional space that grows with the input size.

# Question 3

Given a m*n matrix grid which is sorted in non-increasing order both row-wise and column-wise, return the number of negative number in grid.

Example:

Input: grid = [[4, 3, 2, 1], [3, 2, 1, -1], [-1, -1, -2, -3]]

Output: 8

Exploration: There are 8 negative numbers in the matrix

## Solution

To count the number of negative numbers in a matrix that is sorted in non-increasing order both row-wise and column-wise, we can utilize the property of the matrix and optimize our search.

Here's the step-by-step approach to solve the problem:

1. Initialize a variable count to 0 to store the count of negative numbers.

2. Start from the top-right corner of the matrix, which is grid[0][n-1].

3. Initialize two variables, row and col, to 0 and n-1 respectively, where n is the number of columns in the matrix.

4. Repeat the following steps until row is less than the number of rows (m) and col is greater than or equal to 0:

    - If the element at grid[row][col] is negative, increment count by the number of elements in the current column, which is row + 1. This is because all the elements below the current element in the column will also be negative due to the non-increasing order.

    - Decrement col by 1 to move to the left column.

    - If the element at grid[row][col] is non-negative, increment row by 1 to move to the next row.

5. Return the value of count.

Here's the Python code that implements this approach:

```python
def countNegatives(grid):
    m = len(grid)
    n = len(grid[0])
    count = 0
    row = 0
    col = n - 1

    while row < m and col >= 0:
        if grid[row][col] < 0:
            count += row + 1
            col -= 1
        else:
            row += 1

    return count
```

### Test Cases

```python
# Test Case 1
grid = [[4, 3, 2, 1], [3, 2, 1, -1], [-1, -1, -2, -3]]
result = countNegatives(grid)
print(result)
```

### Conclusion

The countNegatives function takes the matrix as input and initializes the variables m and n to the number of rows and columns respectively. It also initializes count to 0, row to 0, and col to n - 1. Then, it starts from the top-right corner of the matrix and iterates through the matrix using a while loop. If the element at the current position is negative, it increments count by the number of elements in the current column (row + 1) and moves to the left column by decrementing col. If the element is non-negative, it moves to the next row by incrementing row. Finally, it returns the value of count.

The **time complexity of this approach is O(m + n)**, where m is the number of rows and n is the number of columns in the matrix. We move either left or down in each iteration, and the worst case scenario is when we traverse the entire matrix. **The space complexity is O(1)** since we are using only a constant amount of additional space.

3. Return the value of max_wealth.

Here's the Python code that implements this approach:

```python
def maximumWealth(accounts):
    max_wealth = 0

    for customer in accounts:
        wealth = 0
        for amount in customer:
            wealth += amount

        max_wealth = max(max_wealth, wealth)

    return max_wealth
```

### Conclusion

The maximumWealth function takes the accounts matrix as input. It initializes max_wealth to 0 and then iterates over each customer row in the matrix. For each customer, it initializes wealth to 0 and iterates over each amount in the current customer row, adding it to the wealth. After calculating the wealth of the current customer, it updates max_wealth if the new wealth is greater. Finally, it returns the value of max_wealth.

The time complexity of this approach is O(m * n), where m is the number of rows and n is the number of columns in the matrix. We need to visit each element in the matrix once. The space complexity is O(1) since we are using only a constant amount of additional space.
