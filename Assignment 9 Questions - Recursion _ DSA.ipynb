# Question 1

ðŸ’¡ Given an integerÂ `n`, returnÂ *`true`Â if it is a power of two. Otherwise, returnÂ `false`*.

An integerÂ `n`Â is a power of two, if there exists an integerÂ `x`Â such thatÂ `n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

```python
def isPowerOfTwo(n):
    if n <= 0:
        return False
    return (n & (n - 1)) == 0
```

```python
# Test Case 1
n = 1
print(isPowerOfTwo(n))
```

```python
# Test Case 2
n = 16
print(isPowerOfTwo(n))
```

```python
# Test Case 3
n = 3
print(isPowerOfTwo(n))
```

# Question 3

ðŸ’¡ Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24



```python
def factorial(N):
    result = 1
    for i in range(1, N+1):
        result *= i    
    return result
```

```python
# Test Case 1
N = 5
print(factorial(N))
```

```python
# Test Case 2
N = 4
print(factorial(N))
```

# Question 6

ðŸ’¡ Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

```python
def FindNTerm(a, d, N):
    nth_term = a + (N - 1) * d
    return nth_term
```

```python
# Test Case 1
a = 2
d = 1
N = 5
print("The 5th term of the series is:", FindNTerm(a, d, N))
```

```python
# Test Case 2
a = 5
d = 2
N = 10
print("The 10th term of the series is:", FindNTerm(a, d, N))
```

# Question 7

ðŸ’¡ Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = â€œABCâ€*

***Output:***

*â€œABCâ€, â€œACBâ€, â€œBACâ€, â€œBCAâ€, â€œCBAâ€, â€œCABâ€*

**Example 2:**

***Input:***

*S = â€œXYâ€*

***Output:***

*â€œXYâ€, â€œYXâ€*

```python
from itertools import permutations

def print_permutations(S):
    perms = permutations(S)
    for perm in perms:
        print(''.join(perm))
```

```python
# Test Case 1
S = "ABC"
print_permutations(S)
```

```python
# Test Case 2
S = "XY"
print_permutations(S)
```

# Question 8

ðŸ’¡ Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}

Output : 120

**Example 2:**

Input  : arr[] = {1, 6, 3}

Output : 18



```python
def ProductOfArray(arr):
    product = 1
    for num in arr:
        product *= num
    return product
```

```python
# Test Case 1
arr = [1, 2, 3, 4, 5]
print(ProductOfArray(arr))
```

```python
# Test Case 2
arr = [1, 6, 3]
print(ProductOfArray(arr))
```
